## **Problema 1: Calculadora de Permutaciones**

### Descripción matemática:
El primer problema nos relata que dado dos enteros $n$ y $r$, podamos  calcular el número de formas de en las que se puede ordenar r objetos distintos tomados de un conjunto de n objetos distintos.

### Formula utilizada:

La fórmula matemática para calcular el número de formas de ordenar $r$ objetos tomados de un conjunto de $n$ objetos distintos lo cual es la de permutacion es:

$$P(n,r)=\frac{n!}{(n-r)!}$$

## **Problema 2: Calculadora de Combinaciones**

### Descripción matemática:
El segundo problema nos habla que dado dos enteros $n$ y $r$, podamos  calcular el número de formas de en las que se puede ordenar r objetos distintos tomados de un conjunto de n objetos distintos pero sin importar el orden.

### Formula utilizada:

La fórmula matemática para calcular el número de formas de escoger $r$ objetos entre $n$ objetos distintos sin importar el orden (combinaciones) es:

$$\binom{n}{r}=\frac{n!}{r!(n-r)!}$$

```python
## establecemos la permutacion
def calcular_factorial(n):
  if n == 0:
      return 1

  inicio_cadena = 1

  for i in range (1, n + 1):
    inicio_cadena *= i;
  return inicio_cadena

## version alternativa recursiva
def factorial_recursivo(n):
  if n == 0 or n == 1:
    return 1
  return n * factorial_recursivo(n-1)

## funcion principal de la permutacion problema 1
def calcular_permutacion(n, r):
  if n < 0 or r < 0:
    print('Ninguno de los datos puede ser negativo')
    return 0
  elif r > n:
    print('No se puede ordenar mas objetos de los totales')
    return 0
  print(f"Procedimiento matematico: P({n},{r}) = {n}! / ({n}-{r})!")

  fact_n = calcular_factorial(n)
  fact_n_r = calcular_factorial(n - r)

  print(f"--> Sustituyendo valores: P({n},{r}) = {fact_n} / {fact_n_r}")

  resultado_permutacion = fact_n // fact_n_r
  return resultado_permutacion
## funcion principal de combinacion problema 2
def calcular_combinacion(n, r):
   if n < 0 or r < 0:
      print('los datos no pueden ser negativos')
      return 0
   elif r > n:
      print('No se puede ordenar mas objetos de los totales')
      return 0
   print(f"Procedimiento matemático: C({n},{r}) = {n}! / ({r}! * ({n}-{r})!)")

   fact_n = calcular_factorial(n)
   fact_r = calcular_factorial(r)
   fact_n_r = calcular_factorial(n - r)
   print(f"--> Sustituyendo valores: C({n},{r}) = {fact_n} / ({fact_r} * {fact_n_r})")

   resultado_combinacion = fact_n // (fact_r * fact_n_r)
   print(f"--> Propiedad simétrica: Se verifica que C({n},{r}) es igual a C({n},{n-r}) = {resultado_combinacion}")

   return resultado_combinacion

def generar_triangulo_pascal(filas):
    print(f"\n=== TRIANGULO DE PASCAL ({filas} FILAS) ===")

    # se crea una lista para ir guardando el texto de cada fila
    lineas_texto = []

    for n in range(filas):
        fila_actual = []
        for r in range (n + 1):
            valor = calcular_factorial(n) // (calcular_factorial(r) * calcular_factorial(n - r))
            fila_actual.append(str(valor))

        # se une los numeros de la fila con espacios
        texto_fila = "   ".join(fila_actual)
        lineas_texto.append(texto_fila)

    # con esto calculamos el ancho de la base de la piramide
    ancho_base = len(lineas_texto[-1])

    # finalmente se imprime cada linea centrandola respecto al ancho de la base
    for linea in lineas_texto:
        print(linea.center(ancho_base))


## Seccion de pruebas automaticas
print("*" * 50)
print("Pruebas del problema 1: Permutaciones")
print("*" * 50)

# 1. Caso pequeño (Fácil de verificar: 5 objetos tomados de a 2)
print("\nPrueba 1 (Caso pequeño): P(5, 2)")
print(f"Resultado: {calcular_permutacion(5, 2)}")

# 2. Caso mediano
print("\nPrueba 2 (Caso mediano): P(10, 3)")
print(f"Resultado: {calcular_permutacion(10, 3)}")

# 3. Caso extremo
print("\nPrueba 3 (Caso extremo n=r): P(4, 4)")
print(f"Resultado: {calcular_permutacion(4, 4)}")

# 4. Caso extremo
print("\nPrueba 4 (Caso extremo r=0): P(5, 0)")
print(f"Resultado: {calcular_permutacion(5, 0)}")

# 5. Caso de Error (r > n)
print("\nPrueba 5 (Error de validación r > n): P(3, 5)")
print(f"Resultado: {calcular_permutacion(3, 5)}")

print("\n" + "*" * 50)
print("Pruebas del problema 2: Permutaciones")
print("*" * 50)

# 1. Caso pequeño
print("\nPrueba 1 (Caso pequeño): C(5, 2)")
print(f"Resultado: {calcular_combinacion(5, 2)}")

# 2. Caso mediano
print("\nPrueba 2 (Caso mediano): C(10, 3)")
print(f"Resultado: {calcular_combinacion(10, 3)}")

# 3. Prueba de Identidad Simétrica
print("\nPrueba 3 (Identidad Simétrica): C(10, 7)")
print(f"Resultado: {calcular_combinacion(10, 7)}")

# 4. Caso extremo
print("\nPrueba 4 (Caso extremo r=0): C(8, 0)")
print(f"Resultado: {calcular_combinacion(8, 0)}")

# 5. Caso de Error Datos negativos
print("\nPrueba 5 (Error de validación negativos): C(-2, 4)")
print(f"Resultado: {calcular_combinacion(-2, 4)}")

print("\n" + "*" * 50)
print("Triangulo de Pascal")
print("*" * 50)
generar_triangulo_pascal(8)


## calcular_factorial(n), calcular_permutacion(n, r) y calcular_combinacion(n, r) tienen una eficiencia de O(n), ya que son lineales
## generar_triangulo_pascal(filas) tiene una eficiencia de O(F^3) es cubico
```
